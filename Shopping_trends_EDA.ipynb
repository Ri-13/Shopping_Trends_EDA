```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv("shopping_trends.csv")
#Checking the top five columns
sales.head()
```

```python
#Checking the bottom five columns
sales.tail()
```

```python
#Checking the shapes of the dataset (row, column)
sales.shape
```

```python
#Data Cleaning
#1. Checking the datatype
sales.info()
```

```python
#2. Checking the missing values
sales.isnull()
```

```python
#3. Checking how many missing values are there.
sales.isnull().sum()
```

```python
# Checking for duplicates
sales.duplicated().sum()
```

```python
#Data Exploration
#1. Summary of the data
sales.describe()
```

```python
#2. Summarize the most popular payment method
sales.groupby('Payment Method')['Payment Method'].count().sort_values(ascending=False)
```

```python
#Visualization 
base_color = sb.color_palette()[0]

method = sales['Payment Method'].value_counts().index
sb.countplot(data = sales, y = "Payment Method", color = base_color, order = method)
```

```python
# The most popular item purchased
sales.groupby('Item Purchased')['Item Purchased'].count().sort_values(ascending=False)
```

```python
base_color2 = sb.color_palette()[6]

item_purchased = sales['Item Purchased'].value_counts().index
sb.countplot(data = sales, y = 'Item Purchased', color = base_color2 , order = item_purchased)
```

```python
#Checking the distinct gender
sales.Gender.unique()
```

```python
# The most popular item purchased by females
sales[(sales['Gender']=='Female')].groupby('Item Purchased')['Item Purchased'].count().sort_values(ascending=False)
```

```python
#Visualize the location that drives majority of the sales

location = sales['Location'].value_counts().head(5).index
sb.countplot(data = sales, y = 'Location', color = base_color, order = location)
```

```python
# How much do customers spend in average - median and mean

median_purchase = round(sales['Purchase Amount (USD)'].median(),2)
mean_purchase = round(sales['Purchase Amount (USD)'].mean(),2)

print('Median Purchase = '+str(median_purchase))
print('Mean Purchase = '+str(mean_purchase))
```
